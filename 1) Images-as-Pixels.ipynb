```python
import numpy as np
import matplotlib.image as mpimg 
import matplotlib.pyplot as plt
import cv2  
%matplotlib inline
```

```python

image = mpimg.imread('Pic.jpg')

print('Image dimensions color:', image.shape)

gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

print('Image dimensions gs:', gray_image.shape)

plt.imshow(gray_image, cmap='gray')
plt.axis('on')
plt.show()
```

```python
x = 100
y = 300

print(gray_image[x,y])
```

```python
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print('Max: ', max_val)
print('Min: ', min_val)
```

```python
tiny_image = np.array([[0, 20, 30, 150, 120],
                      [200, 200, 250, 70, 3],
                      [50, 180, 85, 40, 90],
                      [240, 100, 50, 255, 10],
                      [30, 0, 75, 190, 220]])

plt.matshow(tiny_image, cmap='gray')
plt.axis('off')
plt.show()
```

```python
tiny_smiley = np.array([
    [0,   0, 0, 0,   0],  # top row
    [0,   255,   0, 255,   0],  # eyes
    [0,   255,   0, 255,   0],  # eyes continue
    [0,     0,   0,   0,   0],  # empty mouth space
    [0,   255, 255, 255,   0]   # smiling mouth
])

plt.matshow(tiny_smiley, cmap='gray')
plt.show()
```
