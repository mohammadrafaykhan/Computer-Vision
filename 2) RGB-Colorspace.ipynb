```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
image = mpimg.imread('Pic.jpg')
plt.imshow(image)
plt.axis('on')
plt.show()
print('Image dimensions color:', image.shape)
```

```python
# Isolate RGB channels
r = image[:,:,0]
g = image[:,:,1]
b = image[:,:,2]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))

ax1.set_title('R channel')
ax1.imshow(r)
ax1.axis('off')

ax2.set_title('G channel')
ax2.imshow(g)
ax2.axis('off')

ax3.set_title('B channel')
ax3.imshow(b)
ax3.axis('off')

plt.show()

## Which area has the lowest value for red? What about for blue?
min_pos_r = np.unravel_index(np.argmin(r), r.shape)
max_pos_r = np.unravel_index(np.argmax(r), r.shape)

print("Red channel min value at:", min_pos_r)
print("Red channel max value at:", max_pos_r)

min_pos_b = np.unravel_index(np.argmin(b), b.shape)
max_pos_b = np.unravel_index(np.argmax(b), b.shape)

print("Blue channel min value at:", min_pos_b)
print("Blue channel max value at:", max_pos_b)

min_pos_g = np.unravel_index(np.argmin(g), g.shape)
max_pos_g = np.unravel_index(np.argmax(g), g.shape)

print("Green channel min value at:", min_pos_g)
print("Green channel max value at:", max_pos_g)
```
